```python
import pandas as pd
import random

pd_precios= pd.read_csv("Matriz_precios_back.csv")
pd_categorias = pd.read_csv("KMEAN_5K_V1")
```

```python
pd_categorias.head()
pd_categoria_0 = pd_categorias[pd_categorias['Categoria']==0]
pd_categoria_1 = pd_categorias[pd_categorias['Categoria']==1]
pd_categoria_2 = pd_categorias[pd_categorias['Categoria']==2]
pd_categoria_3 = pd_categorias[pd_categorias['Categoria']==3]
pd_categoria_4 = pd_categorias[pd_categorias['Categoria']==4]


l_r0=[]
l_r1=[]
l_r2=[]
l_r3=[]
l_r4=[]
for i in range(4):
    r_0 = random.randint(0,int(len(pd_categoria_0))-1)
    r_1 = random.randint(0,int(len(pd_categoria_1))-1)
    r_2 = random.randint(0,int(len(pd_categoria_2))-1)
    r_3 = random.randint(0,int(len(pd_categoria_3))-1)
    r_4 = random.randint(0,int(len(pd_categoria_4))-1)

    l_r0.append(r_0)
    l_r1.append(r_1)
    l_r2.append(r_2)
    l_r3.append(r_3)
    l_r4.append(r_4)
    
    
    
```

```python
import matplotlib.pyplot as plt

# Plot
for indice in l_r0:
    

    plt.plot(pd_precios[ pd_categoria_0.iloc[indice][1] ], marker='o')

    # Labelling 

    plt.xlabel("Fecha")
    plt.ylabel("Precio")
    plt.title(indice)

    # Display

    plt.show()

```

```python
figure,axis = plt.subplots(2,2)
x,y=0,0
for indice in l_r0:
    
    axis[x,y].plot(pd_precios[ pd_categoria_0.iloc[indice][1] ], marker='o')
    if x == 0 and y == 0 : y=+1
    elif x == 0 and y == 1 : x=+1
    else:  y = 0

# Display

plt.show()
```
