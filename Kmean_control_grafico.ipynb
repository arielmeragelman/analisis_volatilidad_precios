```python
def graficado_kmean(archivo_matriz,archivo_cluster,k):


    import pandas as pd
    import matplotlib.pyplot as plt

    
    
    def muestra_aleatorea(cat):        
        import random
        l_r=[]
        
        for cant_rand in range(4):
            l_r.append(random.randint(0,int(len(cat)-1)))
        
        return l_r

    
    
    
    pd_precios= pd.read_csv(archivo_matriz,index_col=[0])
    pd_categorias = pd.read_csv(archivo_cluster,index_col=[0])
    

    lista_categoria=[]
    
    for i in range(k):
        cat=pd_categorias[pd_categorias['Categoria']==i]
        lista_lr= muestra_aleatorea(cat)
        for indice in lista_lr:
            
            plt.plot(pd_precios[ cat.iloc[indice][0] ], marker='o')
            # Labelling 

            plt.xlabel("% Tiempo Medido")
            plt.ylabel("Variacion % Precio")
            plt.title(indice)

            # Display

        plt.show()
    
        
        
        
        
        
     
    
    
    
    
        
    

    
        
        
    
```

# Analisis con K = 6

```python
listado=graficado_kmean("matriz_variacion_v1.csv","KMEAN_6K_V1.csv",6)    
    
    
```
